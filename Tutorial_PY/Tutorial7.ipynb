# Generating Lists using “List Comprehensions”

```python
li = [3, 6, 2, 7]
```

```python
[elem*2 for elem in li] # [ expression for name in list ]
```

```python
li = [('a', 1), ('b', 2), ('c', 7)]
```

```python
[ n * 3 for (x, n) in li]  # [ expression for name in list ]
```

```python
li = [3, 6, 2, 7, 1, 9]
```

```python
[elem * 2 for elem in li if elem > 4] 
# [ expression for name in list if filter] 
# Only 6, 7, 9 satisfy the filter condition.
```

```python
li = [3, 2, 4, 1]
```

```python
[elem*2 for elem in [item+1 for item in li] ]
# [ expression for name in list ]
#[4, 3, 5, 2] -> [8, 6, 10, 4]
```
