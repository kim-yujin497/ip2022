## Geometric Transformations of Images

Scaling 사진확대

```python
import cv2
import numpy as np

img = cv2.imread('mountain1.jpg')
res = cv2.resize(img, None, fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
cv2.imwrite("mountain11.jpg", res)

#OR

height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
cv2.imwrite("mountain12.jpg", res)
```

Translation

```python
import cv2
import numpy as np

img = cv2.imread('mountain1.jpg',0)
rows,cols = img.shape
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Rotation

```python
img = cv2.imread('mountain1.jpg',0)
rows,cols = img.shape

M = cv2.getRotationMatrix2D((cols/2, rows/2),90,1) # (cols/2, rows/2) -> 이미지 중심을 기준으로 회전
dst = cv2.warpAffine(img,M,(cols,rows))
```

Affine / Perspective Transformation

```python
import matplotlib.pyplot as plt

img = cv2.imread('dsu1.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[233,210],[260,496], [692,534],[694,182]])
pts2 = np.float32([[0,0],[500,0],[0,400],[500,400]])

M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(600,380))

plt.subplot(121), plt.imshow(img), plt.title('Input')
plt.subplot(122), plt.imshow(dst), plt.title('Output')
plt.show()
```

Perspective Transformation

```python
img = cv2.imread('dsu1.jpg')

rows,cols,ch = img.shape
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
